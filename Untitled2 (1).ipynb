```python
# Veri setini içeri aktarıyoruz
import pandas as pd


df = pd.read_csv(r"/content/balanced_smart_manufacturing_data.csv")
print(df.head())
```

```python
# Veri setinin genel bilgilerini ve her sütunun veri türünü gösteriyoruz
print(df.info())
```

```python
#Hedef değişkenin "maintenance_required" sınıf dağılımını inceliyoruz
print(df['maintenance_required'].value_counts())
```

```python
# Kategorik verilerden oluşan 'failure_type' sütununu sayısal verilere dönüştürüyoruz ve orijinal sütunu veri setinden çıkarıyoruz
dataencoded=pd.get_dummies(df['failure_type'],drop_first=True)
dataencoded.columns=["failure_type_1","failure_type_2","failure_type_3","failure_type_4"]
dataencoded=dataencoded.map(lambda x: 1 if x==True else 0)
df.drop('failure_type', axis=1, inplace=True)
df=pd.concat([df,dataencoded],axis=1)
df.info()
```

```python
# Fahrenheit'ten Celsius'a dönüştürüyoruz
df["temperature"] = df["temperature"].map(lambda x: 5/9 * (x - 32))
```

```python
# 'timestamp' sütununu yıl, ay, gün gibi bileşenlere ayırıyoruz
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
df['month'] = df['timestamp'].dt.month
df['year'] = df['timestamp'].dt.year
df.drop('timestamp', axis=1, inplace=True)

```

```python
# Doğruluk oranlarını test etmek  için bir çok modeli tanımlıyoruz
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier

import numpy as np

models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "SVC": SVC(probability=True),
    "Logistic Regression": LogisticRegression(max_iter=1000)
}
```

```python
# Çeşitli sınıflandırma modellerini tanımlıyoruz ve 5 katlı çapraz doğrulama ile değerlendiriyoruz




X = df_model.drop('maintenance_required', axis=1)
y = df_model['maintenance_required']

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    print(f"\nModel: {name}")
    fold = 1
    f1_scores = []

    for train_idx, test_idx in skf.split(X, y):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        report = classification_report(y_test, y_pred, output_dict=True)


        class_1_key = str(y_test.unique()[1]) if len(y_test.unique()) > 1 else str(y_test.unique()[0])

        f1 = report[class_1_key]['f1-score'] if class_1_key in report else 0
        f1_scores.append(f1)


        print(f"\nFold {fold} Classification Report:")
        print(classification_report(y_test, y_pred))
        fold += 1


    print(f"\nOrtalama F1-Score (sınıf 1): {np.mean(f1_scores):.4f}")
```

```python
#F1 skoru en iyi olan modelin seçip  en iyi hiperparametresini seçiyoruz
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import confusion_matrix, classification_report


param_dist = {
    'n_estimators': [50, 100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

rf = RandomForestClassifier(random_state=42)

rf_random = RandomizedSearchCV(estimator=rf, param_distributions=param_dist,
                               n_iter=100, cv=3, verbose=2, random_state=42, n_jobs=-1)

rf_random.fit(X_train, y_train)

print("En iyi hiperparametreler:", rf_random.best_params_)

best_model = rf_random.best_estimator_
y_pred_best = best_model.predict(X_test)


confus = confusion_matrix(y_test, y_pred_best)


print("Karışıklık Matrisi:\n", confus)

print(classification_report(y_test, y_pred_best))
```
